```python
# Task 1: Write a function celsius_to_fahrenheit(celsius) that converts a temperature reading from Celsius to Fahrenheit for the weather reports.
# Task 2: Write a function fahrenheit_to_celsius(fahrenheit) that converts temperature data from the old system (which used Fahrenheit) to Celsius for consistency.
# Task 3: Write a function celsius_to_kelvin(celsius) that converts the temperature to Kelvin for scientific analysis.
# Task 4: Write a main program that simulates the input of temperature data from different sensors and converts it to the required units. The system should allow users (weather analysts) to input a temperature and select the conversion they need, displaying the converted temperature.

def celsius_to_fahrenheit(celsius):
    return round((1.8 * celsius)+32,2)

def fahrenheit_to_celsius(fahrenheit):
    return round((fahrenheit-32)*(5/9),2)

def celsius_to_kelvin(celsius):
    return round(celsius + 273.15,2)

def kelvin_to_celsius(kelvin):
    return round(kelvin - 273.15,2)

def convert_temperature(temperature,unit, converted_unit):
    unit = unit.upper()
    converted_unit = converted_unit.upper()

    if unit == converted_unit:
        return temperature
    
    if unit == "C":
        celsius = temperature
    elif unit == "F":
        celsius = fahrenheit_to_celsius(temperature)
    elif unit == "K":
        celsius = kelvin_to_celsius(temperature)
    else:
        return None

    if converted_unit == "C":
        return celsius
    elif converted_unit == "F":
        return celsius_to_fahrenheit(celsius)
    elif converted_unit == "K":
        return celsius_to_kelvin(celsius)
    else:
        return None
    
def check_allowed_units(unit,units):
    return unit in units
    

try:
    allowed_units = ["K","F","C"]
    temperature = float(input("Input your temperature"))
    unit = input("Please enter your current temperature unit:\n• C for Celsius (e.g., 25°C)\n• F for Fahrenheit (e.g., 77°F)\n• K for Kelvin (e.g., 298K)\nYour choice: ").upper()
    check_unit = check_allowed_units(unit,allowed_units)
    if not check_unit:
        print("Only Input Allowed Unit!")
    else:
        converted_unit = input("Convert to which unit? (C/F/K): ").upper()
        check_converted_unit = check_allowed_units(converted_unit, allowed_units)
        if not check_converted_unit:
            print("Only Input Allowed Unit!")
        else:
            converted_temperature = convert_temperature(temperature,unit,converted_unit)
            if converted_temperature is None:
                print("Conversion Error!")
            else:
                print(f"{temperature} {unit} = {converted_temperature} {converted_unit}")
            
except ValueError:
    print("Only Input Numbers!")



```

```python
# You are part of a team developing a text-based adventure game. Players need to interact with the game by entering commands that are strings.

# Task 1: Write a function reverse_command(command) that reverses the player's input command for a hidden puzzle in the game.
# Task 2: Write a function is_palindrome(command) that checks if a command is a palindrome, which unlocks a secret door in the game.
# Task 3: Write a function count_vowels(command) that counts the vowels in a player's command, which determines the strength of a magical spell they cast.
# Task 4: Create a main program that simulates a player entering a command. Based on the command, the program should call the appropriate functions to check for puzzles, unlock secrets, or cast spells, displaying the outcomes.

def reverse_command(command):
    return command[::-1]

def is_palindrome(command):
    return command == command[::-1]

def count_vowels(command):
    count  = 0
    vowels = ["a","e","i","o","u"]
    for letters in command.lower():
        if letters in vowels:
            count+=1
    return count


print("Welcome to the text adventure game!")
print("Type any command to cast a spell!")
print("Special commands:")
print("- Type 'reverse' to find a hidden puzzle")
print("- Type 'unlock' to try opening a secret door")
print("- Type 'quit' to exit the game")


while True:
    command = input("Enter your command")
    if command.lower() == "quit":
        print("Thanks for playing!")
        break

    if command.lower() == "reverse":
        type = input("enter your command to be mirrored!")
        reverse_type = reverse_command(type)
        print(f"Congratulations, you unlock the hidden puzzle by playing {reverse_type}")
        continue

    if command.lower() == "unlock":
        type = input("enter your command to unlock a secret door!")
        if is_palindrome(type):
            print("Congratulations, you unlock a secret door!")
        else:
            print("You did not unlock the secret door!")
        continue

    number_vowels = count_vowels(command)
    print(command)
    print(f"The power of your spell is {number_vowels} atk")
```

```python
# Exercise 3: Personal Finance Manager
# You are building a simple personal finance manager to help users keep track of their expenses and manage their budgets.

# Task 1: Write a function add_income(income, total) that adds a new income amount to the total balance.
# Task 2: Write a function subtract_expense(expense, total) that subtracts an expense amount from the total balance.
# Task 3: Write a function calculate_balance(income, expenses) that calculates the remaining balance after accounting for all expenses.
# Task 4: Create a main program that allows users to input their income and expenses, and then use the functions to display their remaining balance. The program should also warn the user if their expenses exceed their income.

def add_income(income,total):
    return income + total

def subtract_income(expense,total):
    return total-expense

def calculate_balance(incomes,expenses):
    total_income = 0
    total_expense = 0
    for income in incomes:
        total_income+=income
    for expense in expenses:
        total_expense+=expense
    total_balance = total_income - total_expense
    return total_balance

income_list = []
expense_list = []
total_balance = 0

while True:
    text = input("Enter All Your Income, type 'done' to end")
    if text == 'done'.lower():
        print('Your List of Income', income_list)
        break
    try:
        income = float(text)
        print("You Entered", income)
        if income < 0:
            print("Please Enter a Positive Amount")
            continue
        total_balance = add_income(income, total_balance)
        income_list.append(income)
    except ValueError:
        print("Please Enter Valid Numbers!")

while True:
    text = input("Enter All Your Expense, type done to end")
    if text == 'done':
        print('Your List of Expenses', expense_list)
        break
    try:
        expense = float(text)
        print("You Entered", expense)
        if expense < 0:
            print("Please Enter a Positive Amount")
            continue
        total_balance = subtract_income(expense, total_balance)
        expense_list.append(expense)
    except ValueError:
        print("Please Enter Valid Numbers!")

final_balance = calculate_balance(income_list, expense_list)
if final_balance > 0:
    print(f"Your Total Balance is {final_balance}")
else:
    print(f"Your Expenses exceeded your Income by {abs(final_balance)}")
# Calculate Balance function in this case is a little bit redundant, since we are storing the total_balance, we can return this directly



        
```

```python
# Exercise 4: Event Planning System
# Scenario: You are helping an event planner organize a series of events. The planner needs to manage guest lists and schedules effectively.

# Task 1: Write a function factorial(n) that calculates the factorial of the number of guests to determine the possible seating arrangements. Use recursion to implement this function.
# Task 2: Write a function fibonacci(n) that determines the sequence of events based on their importance (using the Fibonacci sequence to assign priorities). Use recursion to implement this function.
# Task 3: Write a function total_guests(events) that sums up the total number of guests across multiple events.
# Task 4: Create a main program that allows the event planner to input the number of guests for each event, and then calculates the possible seating arrangements and priorities of the events using the above functions. Display the results to help with planning.



def factorial(n):
    if n < 0:
        return None
    if n == 0 or n == 1:
        return 1
    return n * factorial(n-1)


def fibonacci(n):
    if n <= 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fibonacci(n-1) + fibonacci(n-2)
    
def total_guests(dictionary):
    sum = 0
    for item in dictionary.values():
        sum+=item
    return sum

events = {"Karaoke Challenge": None , "Dance Contest": None, "Sumo Wrestling": None, "Eating Contest": None, "Fireworks Display": None}

guest_per_event = events.copy()
priority_per_event = events.copy()
seating_per_event = events.copy()

for index, event in enumerate(events, start=1):
    while True:
        try:
            guests = int(input(f"input number of guests for {event}"))
            break
        except ValueError:
            print("Please Enter Whole Numbers Only")
    priority = fibonacci(index)
    possible_seating = factorial(guests)
    priority_per_event[event] = priority
    seating_per_event[event] = possible_seating
    guest_per_event[event] = guests

for key, value in guest_per_event.items():
    print(f" Event:{key} - Number of Guests:{value}")
print("\n")
for key, value in priority_per_event.items():
    print(f" Event:{key} - Priority:{value}")
print("\n")
for key, value in seating_per_event.items():
    print(f" Event:{key} - Possible Seating Arrangement:{value}")
print("\n")
total_number_of_guests = total_guests(guest_per_event)
print("Total Number of Guests is",total_number_of_guests)


```

```python
# Exercise 5: Inventory Management System
# Scenario: You are tasked with developing an inventory management system for a small retail store. The store needs to keep track of products, their quantities, and pricing.

# Task 1: Write a function find_max_price(products) that finds the most expensive product in the inventory.
# Task 2: Write a function find_min_stock(products) that identifies the product with the lowest stock, so the store can reorder it.
# Task 3: Write a function average_price(products) that calculates the average price of all products to help with pricing strategies.
# Task 4: Write a function remove_out_of_stock(products) that removes products with zero stock from the inventory list.
# Task 5: Create a main program that simulates the store’s inventory management process. Allow the user (store manager) to input product details, and then use the functions to analyze and manage the inventory, providing insights such as the most expensive item, average pricing, and stock status

import sys

inventory = {}

def find_max_price(products):
    if not products:
        print("No products in inventory.")
        return
    max_price = -1
    max_products = []
    for key, value in products.items():
        if value['price'] > max_price:
            max_price = value['price']
            max_products = [key]
        elif value['price'] == max_price:
            max_products.append(key)
    print(f"Product(s) with maximum price: {max_products} at {max_price}")

def find_min_stock(products):
    if not products:
        print("No products in inventory.")
        return
    min_stock = sys.maxsize
    min_product = []
    for key, value in products.items():
        if value['quantity'] < min_stock:
            min_stock = value['quantity']
            min_product = [key]
        elif value['quantity'] == min_stock:
            min_product.append(key)
    print(f"Product(s) with minimum stock: {min_product} at {min_stock}")

def average_price(products):
    if not products:
        print("No products in inventory.")
        return
    total = 0
    for value in products.values():
        total += value['price']
    average = total / len(products)
    print(f"Average Price is {average:.2f}")

def remove_out_of_stock(products):
    to_remove = []
    for key, value in products.items():
        if value['quantity'] == 0:   
            to_remove.append(key)

    if to_remove:
        print(f"Removed products: {to_remove}")
    else:
        print("No products removed.")

    for key in to_remove:
        products.pop(key)

    print("\n--- Inventory After Removal of 0 Stock Products---")
    for product, details in products.items():
        print(f"{product}: Quantity = {details['quantity']}, Price = {details['price']}")

while True:
     try:
          add_product = int(input("How many products will you add? "))
          break
     except ValueError:
          print("Please enter a whole number.")

for _ in range(add_product):
    while True:
        product = input("Enter your product: ")
        if product == "":
            print("Please enter product.")
            continue
        if product in inventory.keys():
            print("The product exist in the inventory, choose another name!")
            continue
        break

    while True:
        try:
            quantity = int(input("Enter your quantity: "))
            break
        except ValueError:
            print("Please enter valid number.")

    while True:
        try:
            pricing = float(input("Enter your price: "))
            break
        except ValueError:
            print("Please enter valid number.")

    inventory[product] = {"quantity": quantity, "price": pricing}

find_max_price(inventory)
find_min_stock(inventory)
average_price(inventory)
remove_out_of_stock(inventory)
```
